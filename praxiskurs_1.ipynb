#### Start 
Als Beispieldatensatz betrachten wir die Entwckelung der globalen Rohstoffpreise. Die Daten sind aif der [Website](https://www.worldbank.org/en/research/commodity-markets) der Wordbank verfügbar (monatliche Preise).Zum einlesen von Exeldateien und der konvertierung in den Datentyp DataFrame, benötigen wir in Julia zunächst die folgenden beiden packages 



```julia
using DataFrames
using XLSX
```

```julia

file_path = "CMO-Historical-Data-Monthly.xlsx"
sheet_name = "Monthly Prices"

# Read the Excel file
xlsx_data = XLSX.readdata(file_path, sheet_name, "A5:BT766")

col_names = xlsx_data[1, :]
df = DataFrame(xlsx_data[4:end, :], Symbol.(col_names))
first(df,10)
```

Zunächst geben wir der ersten Spalte in dem DataFrame den namen date um in Zukunft einfacher darauf zurückzugreifen und ändern alle fehlenden Werte in dem DataFrame die durch den String "…" oder ".." gekennzeichnet sind in den Datentyp missing um  

```julia
#1
rename!(df, 1 => "date")

#2
function transform_missing(df)
    for col in names(df)
        df[!, col] = ifelse.((df[!, col] .== "…") .| (df[!, col] .== "..") , missing, df[!, col])
    end
end
```

```julia
transform_missing(df)

first(df,10)
```

als letztes wollen wir die Spalte mit den Daten in ein von Julia unterstütztes Datumsformat umwandeln, um später auch die Zeitpunkte sinnvoll miteinander vergleichen zu können. Hierfür verwenden wir das package Date.

```julia
using Dates 

function transform_dates(date)
    date = String.(date)
    new_date = replace(date,"M" => "-")
    return Date(new_date,dateformat"yyyy-mm")
end 

transform!(df, :date  => ByRow(transform_dates),renamecols = false)

first(df,10)
```

Wir können auch gezielt Abfragen an unseren Dataframe stellen un verändern. Als Beispiel lassen wir uns alle Ölpreise für die Monate ausgeben in denen die Sorte Brent unter 100$ lagen und geben diese geordnet nach Datum aus (beginnend mit dem neusten Datum)

```julia
df_oil_brent = sort(subset(df,:("Crude oil, Brent") => x -> x .<= 100),:date,rev =true)

#df_oil_brent2 = filter(row -> isa(row."Crude oil, Brent", Number) && row."Crude oil, Brent" >= 100, df)

#alternativ über piping 

println(first(df_oil_brent,10))
```

da der obige Code kann sehr schnell sehr unübersichtlich werden, sobald man etwas komplexere Abfragen an den Datensatz stellt.
Deswegen empfielt es sich das package Pipe zu benutzen. 

```julia
using Pipe
 
df_oil_brent = @pipe df |> subset(_,:("Crude oil, Brent") => x -> x.<= 100) |> sort(_ ,:date, rev = true) |> println(first(_,10))
#sort(df,:("Crude oil, Brent"),rev = true)
```

1. Gib die zehn höchsten Preise der Kaffeesorte Arabica aus mit dem jeweiligen Datum.

2. Erstellt im Datensatz eine neue Spalte, die das Verhältnis der Preise der Kaffeesorten Arabica und Robusta beinhaltet. Was ist hier das Maximum und das Minium?

3. Bestimmt für Rohölsorten Brent und WTI den Kalendermonat (Januar - Dezember), für den der Preis über die Jahre hinweg den höchsten Durchschnitt Datensatz hat.

Hinweise: 
- mit der Funktion month aus dem package Dates erhält man den Monat aus einem Datum : 2020-02-01 -> 02
- benutze groupby und combine
- aus dem Package Statistics erhaltet ihr eine vordefinierte mean Funktion


```julia
#Lösung 1. 
@pipe df |> sort(_,:("Coffee, Arabica"), rev = true) |> first(_,10) |> select(_,:date,:("Coffee, Arabica"))
```

```julia
#Lösung 2.
relation(x,y) = x./y 

@pipe df |> transform(_,[:("Coffee, Arabica"),:("Coffee, Robusta")] => relation)  |> select(_,:("Coffee, Arabica_Coffee, Robusta_relation")) |> [maximum(eachrow(_)),minimum(eachrow(_))]
#alternativ 
#@pipe df |> transform(_,[:("Coffee, Arabica"),:("Coffee, Robusta")] => relation) |> maximum(eachrow(_[:,"Coffee, Arabica_Coffee, Robusta_relation"]))
```

```julia
#Lösung 3.

using Statistics

df_mean = @pipe df |> transform(_,:date => x-> month.(x),renamecols = false) |> groupby(_,:date) |> combine(_,:("Crude oil, Brent") => mean, :("Crude oil, WTI") => x -> mean(skipmissing(x))) 

```

```julia
println(findmax(df_mean."Crude oil, Brent_mean"))
println(findmax(df_mean."Crude oil, WTI_function"))
```

```julia
using Plots
plotlyjs()
x = df.date
y = df[:,"Crude oil, Brent"]
years = [i for i in range(df.date[1],df.date[end],step = Year(5))]
year_label = ["$year" for year in range(year(df.date[1]),year(df.date[end]),step = 5)]

plot(x,y,
    label = "Crude oil, Brent",
    xlabel = "year",
    ylabel = "price in dollar \$",
    xlims = extrema(x),
    xticks = (years,year_label),
    legend = :outertopright,
    )
```

#### Aufgabe: 
1. Erstelle einen Plot, der 
- die Zeitreihen der Preise für die Rohölsorten Brent, Dubai, WTI und den Average  beinhaltet,
- ab dem Jahr 1980 bis heute läuft und als alle fünf Jahre ein x-label hat, das nur das Jahr (statt dem vollen Datum) darstellt,
- passende Achsen- und Legendenlabels hat,
- ein passendes Theme verwendet.


```julia
df_1980 = @pipe  df |> subset(_ , :date => (x -> year.(x).>1979))

x = df_1980[:,"date"]

#x2 = df_1980.date
y1=df_1980[:,"Crude oil, Brent"]
y2=df_1980[:,"Crude oil, Dubai"]
y3=df_1980[:,"Crude oil, WTI"]

y = Matrix(select(df_1980,"Crude oil, Brent","Crude oil, Dubai","Crude oil, WTI"))

years = [i for i in range(df_1980.date[1],df_1980.date[end],step = Year(5))]
year_label = ["$year" for year in range(year(df_1980.date[1]),year(df_1980.date[end]),step = 5)]


plot1 = plot(x,y,
        title = "Rohölpreise",
        labels = ["Crude oil, Brent" "Crude oil, Dubai" "Crude oil, WTI"],
        xlabel = "Jahr",
        ylabel = "Preis in \$/mt",
        xlims = extrema(x),
        xticks = (years,year_label),
        #palette = cgrad(:matter, 3),
        #background_colour = :lightgray)
)
```

2. Erstelle einen zweiten Plot der die Differenz der Rohölsorten darstellt, beschrifte beide Plots sinnvoll und ordne diesen untereinander in einem einzigen Plot an. 

```julia
y_dif = y .- df_1980[:,"Crude oil, average"]

plot2 = plot(x,y_dif,
            title = "Preisdifferenz zum Durchschnitt",
            label = ["Crude oil, Brent" "Crude oil, Dubai" "Crude oil, WTI"],
            xlabel = "Jahr",
            ylabel = "Preis in \$/mt",
            xlims = extrema(x),
            xticks = (years,year_label),
            #palette = cgrad(:matter, 3),
            )

plot(plot1,plot2,layout = (2,1),legend = :outertopright)
```
